## PAGINATION

```python
import requests
from bs4 import BeautifulSoup
import re

root = "https://subslikescript.com"
links = []

# Get the first page of movies starting with 'A'
website = f"{root}/movies_letter-A"
result = requests.get(website)
content = result.text
soup = BeautifulSoup(content, 'html.parser')

# Get pagination to determine the number of pages (if needed)
pagination = soup.find("ul", {"class": "pagination"})
pages = pagination.find_all('li', {"class": 'page-item'}) if pagination else []
last_page = int(pages[-2].text) if pages else 1

# Loop through the first page (or all pages with range(1, int(last_page)+1))
for page in range(1, 2):
    website = f"{root}/movies_letter-A?page={page}"
    result = requests.get(website)
    content = result.text
    soup = BeautifulSoup(content, 'html.parser')
    box = soup.find('article', {"class": "main-article"})

    # Collect all links to movie transcripts
    for link in box.find_all("a", href=True):
        links.append(link["href"])

# Function to sanitize filenames
def sanitize_filename(title):
    return re.sub(r'[<>:"/\\|?*]', '', title)

# Visit each movie transcript link and save the content
for link in links:
    try:
        website = f"{root}/{link}"
        result = requests.get(website)
        content = result.text
        soup = BeautifulSoup(content, 'html.parser')
        box = soup.find('article', {"class": "main-article"})
        title = box.find("h1").get_text()
        transcript = box.find("div", {"class": "full-script"}).\
        get_text(strip=True, separator=" ")

        filename = sanitize_filename(title) + ".txt"
        with open(filename, "w", encoding="utf-8") as file:
            file.write(transcript)

        print(f"Saved: {filename}")

    except Exception as e:
        print("Link not working:", link)
        print("Error:", e)
```

## FLIPKART

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

# Specify the URL of the HTML file to be read
url = "flipkart.html"

# Open the file in read mode and specify the encoding as UTF-8
with open(url, "r", encoding="utf-8") as file:
    # Read the contents of the file
    html_content = file.read()

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Find all the div elements with class "_2kHMtA"
product_divs = soup.find_all("div", class_="_2kHMtA")

# Get the number of div elements found
print("Number of products found:", len(product_divs))

# Initialize empty lists to store mobile names, ratings, first prize, and final price
name_mobile = []
rating = []
fprize_mobile = []
prize_mobile = []

# Iterate over each div element in the soup object
for i in product_divs:
    # Extract the name of the mobile and add it to the list
    name_mobile.append(i.find("div", class_="_4rR01T").text)

    # Extract the rating, convert to float, and add to the list
    rating.append(float(i.find("div", class_="_3LWZlK").text))

    # Extract the first price, clean and convert to integer, then add to the list
    first_price_text = i.find("div", class_="_30jeq3 _1_WHN1").text
    fprize_mobile.append(int(first_price_text[1:3] + first_price_text[4:]))

    try:
        # Try to extract the original price if available
        old_price_text = i.find("div", class_="_3I9_wc _27UcVY").text
        prize_mobile.append(int(old_price_text[1:].replace(",", "")))
    except:
        # If not found, add NaN
        prize_mobile.append(np.nan)

# Create a dictionary to store the extracted data
data = {
    "name_mobile": name_mobile,
    "rating": rating,
    "fprize_mobile": fprize_mobile,
    "prize_mobile": prize_mobile
}

# Create a DataFrame using the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
df
```

## POLITIFACT

```python
#Import the dependencies
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.request
import time

#Create lists to store the scraped data
authors = []
dates = []
statements = []
sources = []
targets = []

#Create a function to scrape the site
def scrape_website(page_number, source):
    page_num = str(page_number) #Convert the page number to a string

    '''source: a certain speaker only'''
    URL = 'https://www.politifact.com/factchecks/list/?\
    page={}&speaker={}'.format(page_num, source)

    '''source: all'''
    # URL = 'https://www.politifact.com/factchecks/list/?page='+page_num #append the page number to complete the URL

    webpage = requests.get(URL)  #Make a request to the website
    #time.sleep(3)
    soup = BeautifulSoup(webpage.text, "html.parser") #Parse the text from the website
    #Get the tags and it's class
    statement_footer =  soup.find_all('footer',attrs={'class':'m-statement__footer'})  #Get the tag and it's class
    statement_quote = soup.find_all('div', attrs={'class':'m-statement__quote'}) #Get the tag and it's class
    statement_meta = soup.find_all('div', attrs={'class':'m-statement__meta'})#Get the tag and it's class
    target = soup.find_all('div', attrs={'class':'m-statement__meter'}) #Get the tag and it's class
    #loop through the footer class m-statement__footer to get the date and author
    for i in statement_footer:
        link1 = i.text.strip()
        name_and_date = link1.split()
        first_name = name_and_date[1]
        last_name = name_and_date[2]
        full_name = first_name+' '+last_name
        month = name_and_date[4]
        day = name_and_date[5]
        year = name_and_date[6]
        date = month+' '+day+' '+year
        dates.append(date)
        authors.append(full_name)
    #Loop through the div m-statement__quote to get the link
    for i in statement_quote:
        link2 = i.find_all('a')
        statements.append(link2[0].text.strip())
    #Loop through the div m-statement__meta to get the source
    for i in statement_meta:
        link3 = i.find_all('a') #Source
        source_text = link3[0].text.strip()
        sources.append(source_text)
    #Loop through the target or the div m-statement__meter to get the facts about the statement (True or False)
    for i in target:
        fact = i.find('div', attrs={'class':'c-image'}).\
        find('img').get('alt')
        targets.append(fact)

#Loop through 'n-1' webpages to scrape the data
n=2
for i in range(1, n):
    scrape_website(i, source='joe-biden')

#Create a new dataFrame
data = pd.DataFrame(columns = ['author',  'statement', \
                               'source', 'date', 'target'])
data['author'] = authors
data['statement'] = statements
data['source'] = sources
data['date'] = dates
data['target'] = targets
#Show the data set
print(data)
data.iloc[:5].to_csv('politifact-joe-biden-sample.csv', \
                     index=False, sep=',')
```

## rottentomatoes

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "30 Most Popular Movies Right Now： What to Watch In Theaters and Streaming ｜ Rotten Tomatoes (5_28_2025 8:32:40 AM).html"
# Open the file in read mode and specify the encoding as UTF-8
with open(url, "r", encoding="utf-8") as file:
    # Read the contents of the file
    html_content = file.read()

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

movie_containers = soup.find_all("div", class_="row countdown-item")

titles = []
ratings = []
years = []

for container in movie_containers:
    # Extract the movie title from <h2><a>
    title_tag = container.find("h2").find("a")
    title = title_tag.text.strip() if title_tag else "N/A"

    # Extract the rating
    rating_tag = container.find("span", class_="tMeterScore")
    rating = rating_tag.text.strip() if rating_tag else "N/A"

    # Try to extract the year from a nearby <span> (likely in the synopsis area)
    year_tag = container.find("span", class_="start-year")
    if not year_tag:
        # Alternative: try to extract from synopsis area using parentheses
        synopsis = container.find("div", class_="synopsis")
        year_text = synopsis.text if synopsis else ""
        start = year_text.find("(")
        end = year_text.find(")", start)
        year = year_text[start+1:end-1] if start != -1 and end != -1 else "N/A"
    else:
        year = year_tag.text.strip()
    year = year.replace("(", "").replace(")", "")

    titles.append(title)
    ratings.append(rating)
    years.append(year)

# Create DataFrame
df = pd.DataFrame({
    "Movie Title": titles,
    "Year": years,
    "Rating": ratings
})

print(df)
```

```python
import requests
from bs4 import BeautifulSoup

def extract_urls(url):
    """Extracts all the URLs from a given URL."""
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all the li elements.
    li_elements = soup.find_all("li")
    
    # Extract the href attribute from each li element.
    urls = []
    for li_element in li_elements:
        href = li_element.find("a", href=True)
        if href is not None:
            print(href["href"])
            urls.append(href["href"])

    return urls

if __name__ == "__main__":
    url = "https://www.python.org/"
    urls = extract_urls(url)
    # Optionally print the list of URLs
    print(urls)
```
